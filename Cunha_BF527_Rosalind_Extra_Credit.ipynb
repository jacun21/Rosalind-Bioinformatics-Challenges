## Challenge 3: Reverse complementing a DNA strand (ID: REVC)



```python

def run(ss): #x is an input of DNA bases in string form

    comp = '' #initializing an empty string to store complementary bases 

    ss = ss.upper() #applying upper() to input DNA strand to make sure they're all upper case
    
    #Loop that goes through sequence, adds complementary base to comp string
    for base in ss:        
        if base == 'A':
            comp += ('T')
        elif base == 'T':
            comp += ('A')
        elif base == 'G':
            comp += ('C')
        elif base == 'C':
            comp += ('G')


    print(comp[::-1]) #returning the reverse complement (takes the complementary strand, reads it backwards)
        
```

## Challenge 4: Rabbits and Recurrence Relations (ID: FIB)

```python

#Useful article that helped me implement this code: https://realpython.com/fibonacci-sequence-python/

  # print out the number of rabbit pairs that exist after n months if each
  # pair of living rabbits producs a litter of k rabbit pairs: https://realpython.com/fibonacci-sequence-python/#using-iteration-and-a-python-function

def run(n,k):
    
    #handling base cases
    if n == 1:
        print(1)
    elif n == 2:
        print(2)
    
    previous, fib_number = 1,1 #initializing previous and fib_number
    
    for i in range(3, n+1):
        
        new_pair = previous * k
        previous, fib_number = fib_number, fib_number + new_pair
        
    print(fib_number)
```

## Challenge 5: Computing GC Content (ID: GC)



```python

def run(fasta_str) :
  # calculate the %GC content of each DNA sequence in the fasta records of fasta_str
  # fasta_str is a single string with each line separated by a newline character
  # retain the sequence name and gc content of the sequence with the highest %gc
  # print out the sequence name with the print function on its ownline
  # then print out the %GC of the sequence with the print function on the next line
    
    max_gc_percent = 0
    max_gc_seq_id = ""
    
    fh = fasta_str.strip().split("\n")
    


    current_seq_id = ""
    current_seq = ""

    for line in fh:
        line = line.strip()

        if line.startswith('>'):
            #processing previous sequence

            if current_seq_id and current_seq:
                counts = {'A': 0, 'T': 0, 'G': 0, 'C': 0} #dictionary to store counts of each nucleotide type

                #counting the base in the counts dictionary
                for base in current_seq:
                    counts[base] += 1


                #Compute GC content
                gc = (counts['G'] + counts['C']) / sum(counts.values()) * 100

                if gc > max_gc_percent:

                    max_gc_percent = gc
                    max_gc_seq_id = current_seq_id


        #processing new sequence
            current_seq_id = line[1:]
            current_seq = ""
        else:
            current_seq += line

    #process last sequence

    if current_seq_id and current_seq:

        counts = {'A': 0, 'T': 0, 'G': 0, 'C': 0} #dictionary to store counts of each nucleotide type

        #counting the base in the counts dictionary
        for base in current_seq:
            counts[base] += 1


        #Compute GC content
        gc = (counts['G'] + counts['C']) / sum(counts.values()) * 100

        if gc > max_gc_percent:

            max_gc_percent = gc
            max_gc_seq_id = current_seq_id

    print(max_gc_seq_id)
    print(round(max_gc_percent, 6))

        
                        
    
```

## Challenge 6: Counting point mutations (ID: HAMM)


```python

def run(s,t) :
  # count the number of base positions that do not have the same nucleotide
  # across DNA sequences s and t
  # print out the number of differing positions with the print function
    
    hamming_distance = 0
    
    for i in range(len(s)):
        
        if s[i] != t[i]:

            hamming_distance += 1
                
        else:
            continue
                
    print(hamming_distance)
```

## Challenge 7: Reproducing Mendel's First Law (ID: IPRB)


```python

#This video helped me implement this solution: https://www.youtube.com/watch?v=8X7WNs6R2zQ


import numpy as np

#creating arrays to represent homozygous dominant, heterozygous, homozygous recessive alleles

AA = np.array([[0,0]])
Aa = np.array([[0,1]])
aa = np.array([[1,1]])

#lists to store probabilities of each outcome, fraction of each dominant offspring, and sum of probabilities

each_probability = []
each_dominant_fraction = []
sum_probabilities = []


def run(k,m,n):
    
    organisms = [k,m,n] #array representing each of the organism types
    
    #loop that iterates through possible pairings, calculates the probabilities of getting each offspring type, adding those probabilities to a list
    
    for i in range(len(organisms)):
        for j in range(len(organisms)):
            add_probs = k+m+n
            
            if i != j:
                probability = (organisms[i]/add_probs) * ((organisms[j])/(add_probs - 1))
            
            else:
                probability = (organisms[i]/add_probs) * ((organisms[j] - 1)/(add_probs - 1))
            
            each_probability.append(probability)
            
            
     
    organism_genotypes = [AA, Aa, aa] #create list to store possible genotypes
    
    #Loop that calculates the fraction of offspring that are expected to have the dominant phenotype
    
    for k in organism_genotypes:
        for l in organism_genotypes:
            matrix = k.T * l
            
            dominant_fraction = np.sum(matrix == 0) / 4
            
            each_dominant_fraction.append(dominant_fraction)
            
    
    
    #final loop that goes through each each_dominant_fraction and each_probability lists, multiplies the values together, and adds them to sum_probabilities list. Prints result 
    
    for m in range(0,9):
        sum_probabilities.append(each_dominant_fraction[m] * each_probability[m])
    print(sum(sum_probabilities))
            
    
    
```

## Challenge #8: Translating RNA into Protein (ID: PROT)

```python

#I originally wrote this script for a BF550 project 


def rna_to_aa(rna):

    #Dictionary with all possible codons as keys and amino acid symbols as values.
    #Stop codons are asterisks (*).

    codons = {"UUU":"F", "UUC":"F", "UUA":"L", "UUG":"L",
           "UCU":"S", "UCC":"S", "UCA":"S", "UCG":"S",
           "UAU":"Y", "UAC":"Y", "UAA":"", "UAG":"",
           "UGU":"C", "UGC":"C", "UGA":"", "UGG":"W",
           "CUU":"L", "CUC":"L", "CUA":"L", "CUG":"L",
           "CCU":"P", "CCC":"P", "CCA":"P", "CCG":"P",
           "CAU":"H", "CAC":"H", "CAA":"Q", "CAG":"Q",
           "CGU":"R", "CGC":"R", "CGA":"R", "CGG":"R",
           "AUU":"I", "AUC":"I", "AUA":"I", "AUG":"M",
           "ACU":"T", "ACC":"T", "ACA":"T", "ACG":"T",
           "AAU":"N", "AAC":"N", "AAA":"K", "AAG":"K",
           "AGU":"S", "AGC":"S", "AGA":"R", "AGG":"R",
           "GUU":"V", "GUC":"V", "GUA":"V", "GUG":"V",
           "GCU":"A", "GCC":"A", "GCA":"A", "GCG":"A",
           "GAU":"D", "GAC":"D", "GAA":"E", "GAG":"E",
           "GGU":"G", "GGC":"G", "GGA":"G", "GGG":"G",}
    
    aa_seq = '' #empty string that will store amino acid symbols
    
    if len(rna)%3 == 0:  #If the RNA strand is divisible by 3
        
        for i in range(0,len(rna),3):
            
            three_bases = rna[i:i+3] #checks current base and the two next to it, see which codon it is
            
            aa_seq += codons[three_bases] #adds the amino acid symbol for the current codon to the aa_seq string
            
            
            
    else: #else if the input RNA sequence is not perfectly divisible by 3
        
        extra_bases = len(rna)%3 #find how many extra bases there are at the end
        
        for i in range(0,len(rna)- extra_bases,3): #running same loop as above, just subtracting off the extra bases
            
            three_bases = rna[i:i+3] #checks current base and the two next to it, see which codon it is
            
            aa_seq += codons[three_bases] #adds the amino acid symbol for the current codon to the aa_seq string
            
        
        
    print(aa_seq)

```
